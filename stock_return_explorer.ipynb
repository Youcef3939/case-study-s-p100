```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tickers = ["AAPL", "TSLA"]
data = yf.download(tickers , start="2023-01-01" , end="2024-01-01")["Close"]
```

```python
returns = data.pct_change().dropna()
```

```python
summary = pd.DataFrame({
    "Mean Returns": returns.mean(),
    "Std Dev": returns.std(),
    "Min Return": returns.min(),
    "Max Returns": returns.max(),
})
print(summary)
```

```python
data.plot(title="Stock Prices")
plt.show()
```

```python
returns.hist(bins=50 , figsize=(10,4))
plt.suptitle("Distribution of Returns")
plt.show()
```
